```python
# ===== Titanic: Variables numéricas (hist, boxplot, correlación, heatmap) =====
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Cargar datos
try:
    Titanic  # si ya existe el DataFrame, no hagas nada
except NameError:
    # ajusta el nombre/ubicación del archivo si es necesario
    Titanic = pd.read_csv("Titanic.csv")

# 2) Seleccionar columnas numéricas de interés
num_cols = ['Age', 'Fare', 'SibSp', 'Parch']
num_cols = [c for c in num_cols if c in Titanic.columns]  # por si falta alguna

# 3) Histogramas
for col in num_cols:
    plt.hist(Titanic[col].dropna(), bins=10, edgecolor='black')
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

# 4) Boxplots
for col in num_cols:
    plt.boxplot(Titanic[col].dropna())
    plt.title(f'Boxplot de {col}')
    plt.ylabel(col)
    plt.show()

# 5) Matriz de correlación (solo numéricas disponibles)
corr = Titanic[num_cols].corr()
print("Matriz de correlación:\n", corr, "\n")

# 6) Heatmap de correlación
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt=".2f",
            linewidths=.5, cbar_kws={"shrink": .8})
plt.title('Mapa de calor de correlación - Variables Numéricas')
plt.tight_layout()
plt.show()
# ===== fin =====
```
